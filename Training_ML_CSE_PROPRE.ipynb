```python
# Spliter les DATES
i = 0 
while i < len(date_columns):
    column_targeted = date_columns[i]
    year = pd.DatetimeIndex(df[column_targeted]).year
    new_column = df.columns.get_loc(column_targeted) +1
    column_name = column_targeted+"YEAR"
    df.insert(new_column,column_name, year, allow_duplicates=False)
    i = i+1
len(df.columns)
```

```python
# Calculer Age
Year_list =[]
i = 0 
while i < len(df.inf_birthday):
    Year_targeted_1 = df.inf_birthdayYEAR[i]
    Year_targeted_2 = df.inf_decisiondateYEAR[i]
    year = Year_targeted_2-Year_targeted_1
    Year_list.append(year)
    i = i+1
new_column = df.columns.get_loc('inf_decisiondate') +1
column_name = "Decisiondate_age"
df.insert(new_column,column_name, Year_list, allow_duplicates=False)
len(df.columns)
```
